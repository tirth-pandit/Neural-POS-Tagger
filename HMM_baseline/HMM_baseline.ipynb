```python
import codecs
import os
import sys
import random
import time

import numpy as np
import matplotlib.pyplot as plt 


def plot_histogram(dict_class,heading):
  name = list(dict_class.keys())
  values = list(dict_class.values())
  x = range(len(name))
  fig = plt.figure(figsize = (10, 5))
  
  # creating the bar plot
  plt.bar(name, values, color ='maroon', width = 0.4)
  plt.xticks(x, name, rotation=90)
  plt.ylabel("Count")
  plt.title("Data Set Tag " + str(heading))

def get_train_test_data(path):
  file_path = path
  heading = file_path.split(os.sep)[-1];
  print('Processing ',heading)

  file1 = open(file_path, 'r')
  Lines = file1.readlines()
  
  sentence = []
  for lines in Lines:
    sentence.append(lines)
  
  print("Total Sentences in " + str(heading) + " " + ":" + " " + str(len(sentence)))
  random.shuffle(sentence)

  train_len = int(len(sentence)*0.9)

  train_sentence = sentence[0:train_len]
  test_sentence = sentence[train_len+1:]
  print("Total Train Sentences in " + str(heading) + " " + ":" + " " + str(len(train_sentence)))
  print("Total Test Sentences in " + str(heading) + " " + ":" + " " + str(len(test_sentence)))

  return sentence,train_sentence,test_sentence

def prepare_data_test(sentence,heading):
  ff = open('/content/drive/MyDrive/nlpproject/train_test/test_' + str(heading) + '.txt','a')
  label = []
  for line in sentence:
    temp = line.split()[1:]
    t = []
    for i in temp:
      x = i.split('\\')
      ff.write(x[0] + " ")
      t.append(x[1])
    ff.write("\n")
    label.append(t)
  
  return label
def prepare_data_train(sentence,heading):
  dict_class = {}
  f = open('/content/drive/MyDrive/nlpproject/train_test/train_' + str(heading),'a')
  for line in sentence:
    temp = line.split()[1:]
    f.write("<s> START")
    f.write("\n")
    for i in temp:
      x = i.split('\\')
      f.write(x[0] + " " + x[1])
      if(x[1] not in dict_class.keys()):
        dict_class[x[1]] = 1
      else:
        dict_class[x[1]] += 1
      f.write("\n")
    f.write("</s> END")
    f.write("\n")

  print(dict_class)
  f.close()
  return dict_class
```

# Data Preperation

```python
label_dict = {}

def main():
  for a,b,files in os.walk('/content/drive/MyDrive/nlpproject/data/'):
    for j in files:
      file_path = a+j
      print(file_path)
      heading = file_path.split(os.sep)[-1]
      sentence,train_sentence,test_sentence = get_train_test_data(file_path)
      dict_class = prepare_data_train(train_sentence,heading)
      label = prepare_data_test(test_sentence,heading)
      label_dict[heading] = label
      plot_histogram(dict_class,heading)
      print("------------------------------------------")

main()
```

# Training

```python
import codecs
import os
import sys
import time

tags = ['N_NN', 'PR_PRL', 'JJ', 'V_VAUX_VNP', 'V_VAUX', 'N_NNP', 'RD_PUNC', 'DM_DMD', 'N_NST', 'PR_PRI', 'PR_PRP', 'V_VM', 'PSP', 'DM_DMI', 'RP_RPD', 'QT_QTF', 'RP_NEG', 'CC_CCD', 'QT_QTC', 'RD_SYM', 'CC_CCS', 'PR_PRF', 'PR_PRQ', 'DM_DMQ', 'CC_CCS_UT', 'PR_PRC', 'QT_QTO', 'RB', 'RD_RDF', 'DM_DMR','RP_INJ','RD_ECH','RP_INTF','RP_CL']

def max_connect(x, y, viterbi_matrix, emission, transmission_matrix):
	max = -99999
	path = -1
	
	for k in range(len(tags)):
		val = viterbi_matrix[k][x-1] * transmission_matrix[k][y]
		if val * emission > max:
			max = val
			path = k
	return max, path

def main(filepath,testpath,heading):
  print(heading)
  start_time = time.time()
  languages = heading
  exclude = ["<s>", "</s>", "START", "END"]
  wordtypes = []
  tagscount = []

  f = codecs.open(filepath, 'r', encoding='utf-8')
  file_contents = f.readlines()

  for x in range(len(tags)):
    tagscount.append(0)

  for x in range(len(file_contents)):
    line = file_contents.pop(0).strip().split(' ')
    for i, word in enumerate(line):
      if i == 0:
        if word not in wordtypes and word not in exclude:
          wordtypes.append(word)
      else:
        if word in tags and word not in exclude:
          tagscount[tags.index(word)] += 1
  f.close()

  emission_matrix = []
  transmission_matrix = []
    
  # Initialize emission matrix
  for x in range(len(tags)):
    emission_matrix.append([])
    for y in range(len(wordtypes)):
      emission_matrix[x].append(0)
    
  for x in range(len(tags)):
    transmission_matrix.append([])
    for y in range(len(tags)):
      transmission_matrix[x].append(0)

  f = codecs.open(filepath, 'r', encoding='utf-8')
  file_contents = f.readlines()

  row_id = -1
  for x in range(len(file_contents)):
    line = file_contents.pop(0).strip().split(' ')
    
    if line[0] not in exclude:
      col_id = wordtypes.index(line[0])
      prev_row_id = row_id
      row_id = tags.index(line[1])
      emission_matrix[row_id][col_id] += 1
      if prev_row_id != -1:
        transmission_matrix[prev_row_id][row_id] += 1
    else:
      row_id = -1

  # Divide each entry in emission matrix by appropriate tag count to store probabilities in each entry instead of just count
  for x in range(len(tags)):
    for y in range(len(wordtypes)):
      if tagscount[x] != 0:
        emission_matrix[x][y] = float(emission_matrix[x][y]) / tagscount[x]

  # Divide each entry in transmission matrix by appropriate tag count to store probabilities in each entry instead of just count
  for x in range(len(tags)):
    for y in range(len(tags)):
      if tagscount[x] != 0:
        transmission_matrix[x][y] = float(transmission_matrix[x][y]) / tagscount[x]

  print(time.time() - start_time, "seconds for training")

  start_time = time.time()

  # Open the testing file to read test sentences

  file_test = codecs.open(testpath, 'r', encoding='utf-8')
  test_input = file_test.readlines()

  # Declare variables for test words and pos tags
  test_words = []
  pos_tags = []

  # Create an output file to write the output tags for each sentences
  file_output = codecs.open("/content/drive/MyDrive/nlpproject/output/"+ languages +"_tags.txt", 'w', 'utf-8')
  file_output.close()

  # For each line POS tags are computed
  final_label =[]
  for j in range(len(test_input)):
    
    test_words = []
    pos_tags = []

    line = test_input.pop(0).strip().split(' ')
    
    for word in line:
      test_words.append(word)
      pos_tags.append(-1)

    viterbi_matrix = []
    viterbi_path = []
    
    # Initialize viterbi matrix of size |tags| * |no of words in test sentence|
    for x in range(len(tags)):
      viterbi_matrix.append([])
      viterbi_path.append([])
      for y in range(len(test_words)):
        viterbi_matrix[x].append(0)
        viterbi_path[x].append(0)

    # Update viterbi matrix column wise
    for x in range(len(test_words)):
      for y in range(len(tags)):
        if test_words[x] in wordtypes:
          word_index = wordtypes.index(test_words[x])
          tag_index = tags.index(tags[y])
          emission = emission_matrix[tag_index][word_index]
        else:
          emission = 0.001

        if x > 0:
          max, viterbi_path[y][x] = max_connect(x, y, viterbi_matrix, emission, transmission_matrix)
        else:
          max = 1
        viterbi_matrix[y][x] = emission * max

    maxval = -999999
    maxs = -1
    for x in range(len(tags)):
      if viterbi_matrix[x][len(test_words)-1] > maxval:
        maxval = viterbi_matrix[x][len(test_words)-1]
        maxs = x
      
    for x in range(len(test_words)-1, -1, -1):
      pos_tags[x] = maxs
      maxs = viterbi_path[maxs][x]

    file_output = codecs.open("/content/drive/MyDrive/nlpproject/output/"+ languages +"_tags.txt", 'a', 'utf-8')
    temp_label = []
    for i, x in enumerate(pos_tags):
      file_output.write(test_words[i] + "_" + tags[x] + " ")
      temp_label.append(tags[x])
    file_output.write(" ._.\n")	
    final_label.append(temp_label)

  f.close()
  file_output.close()
  file_test.close()
  print(time.time() - start_time, "seconds for testing 100 Sentences")
  return final_label
```

```python
train = ['/content/drive/MyDrive/nlpproject/train_test/train_guj_entertainment_sample3.txt', \
         '/content/drive/MyDrive/nlpproject/train_test/train_guj_philosophy_sample4.txt',\
         '/content/drive/MyDrive/nlpproject/train_test/train_guj_science and technology_sample6.txt',\
         '/content/drive/MyDrive/nlpproject/train_test/train_guj_sports_sample7.txt',\
         '/content/drive/MyDrive/nlpproject/train_test/train_guj_religion_sample5.txt',\
         '/content/drive/MyDrive/nlpproject/train_test/train_guj_economy_sample2.txt',\
         '/content/drive/MyDrive/nlpproject/train_test/train_guj_art and culture_sample1.txt']

test = ['/content/drive/MyDrive/nlpproject/train_test/test_guj_entertainment_sample3.txt.txt',\
        '/content/drive/MyDrive/nlpproject/train_test/test_guj_philosophy_sample4.txt.txt',\
        '/content/drive/MyDrive/nlpproject/train_test/test_guj_science and technology_sample6.txt.txt',\
        '/content/drive/MyDrive/nlpproject/train_test/test_guj_sports_sample7.txt.txt',\
        '/content/drive/MyDrive/nlpproject/train_test/test_guj_religion_sample5.txt.txt',\
        '/content/drive/MyDrive/nlpproject/train_test/test_guj_economy_sample2.txt.txt',\
        '/content/drive/MyDrive/nlpproject/train_test/test_guj_art and culture_sample1.txt.txt']

heading = ['guj_entertainment_sample3','guj_philosophy_sample4','guj_science and technology_sample6','guj_sports_sample7','guj_religion_sample5','guj_economy_sample2','guj_art and culture_sample1']

def trains():
  label_predict = {}
  for i in range(len(train)):
    label = main(train[i],test[i],heading[i])
    label_predict[heading[i]] = label
    print("--------------------------------------------------------------------------")
    print("\n")
  return label_predict
```
